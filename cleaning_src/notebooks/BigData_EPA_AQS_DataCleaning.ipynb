```python
# read the csv into the stream
datafile = "epa_aqs_air_AQI-POLLUTANTS_2019-2020.csv"
ds = stream(datafile)

#Print the first ten rows of the dataset
ds.head()
```

```python
# Profile the rows using the default profiler
from openclean.profiling.column import DefaultColumnProfiler

profiles = ds.profile(default_profiler=DefaultColumnProfiler)
```

```python
# Print overview of the profiling results
profiles.stats()
```

```python
#Check distinct values
pp.pprint(ds.distinct("validity_indicator"))
pp.pprint(ds.distinct("aqi"))
pp.pprint(ds.distinct("pollutant_standard"))
```

```python
# according to the AQS Data dictionairy(https://aqs.epa.gov/aqsweb/documents/AQS_Data_Dictionary.html)
"""
The Air Quality Index for the day for the pollutant, if applicable. The air quality index is a unitless measure of the amount of pollutant that can be used to relate the pollutant to the healthy levels and indicate possible health concerns with elevated levels. The Air Quality Index (AQI) is a measure for reporting daily air quality. It focuses on health effects that may be experienced within a few hours or days after breathing polluted air. AQI is calculated for the following major air pollutants regulated by the Clean Air Act: Ozone, PM 2.5, PM 10, carbon monoxide, sulfur dioxide, and nitrogen dioxide. The AQI is a mapping from pollutant concentrations to the common index. The index is based on defining seven levels of concentration/index values that are classified as follows:

Good (with AQI values from 0 to 50),
Moderate (with AQI values from 51 to 100),
Unhealthy for Sensitive Groups (with AQI values from 101 to 150),
Unhealthy (with AQI values from 151 to 200),
Very Unhealthy (with AQI values from 201 to 300),
Hazardous (with AQI values above 301)."""
pp.pprint(profiles.minmax("aqi"))
```

```python
# Since the range of values present seems pretty sensible, let's check to see if there is any overlap between
# the rows that don't have it filled in and those that do
from openclean.function.eval.base import Col
empty_aqi = ds.filter(predicate=Col("aqi") == "")
non_empty_aqi = ds.filter(predicate=Col("aqi") != "")
print("Params with empty AQI:")
pp.pprint(empty_aqi.distinct_values('parameter'))
print("Params with non-empty AQI: ")
pp.pprint(non_empty_aqi.distinct_values('parameter'))
```

```python
# There seems to be some overlap, but lets check the sample duration to insure we're not comparing apples to oranges
print("Sample Duration of rows with empty AQI")
pp.pprint(empty_aqi.distinct_values("sample_duration"))
print("Sample Duration of rows with non-empty AQI")
pp.pprint(non_empty_aqi.distinct_values("sample_duration"))

# Since all of the empty AQIs are 1 Hour duration, it seems reasonable to assume that AQI
# is not applicable to those
```

```python
from openclean.operator.map.violations import fd_violations
from openclean.operator.collector.count import distinct
# Let's check the pollutant_standard and parameter columns to make sure those all match

param_violations = fd_violations(ds.to_df(), ["parameter", "parameter_code"], ["pollutant_standard"])
print(f"# of violations for FD(parameter -> pollutant_standard) is {len(param_violations)}")
for key, gr in param_violations.items():
    print(gr[['parameter', "parameter_code", "pollutant_standard"]])

# Since these all seem to line up, we will leave them as is
```
